```python
import pandas as pd
```

```python
df = pd.read_csv('dataReviews.csv')

df.sample()
```

```python
education=['Art&Design','Books&Reference','Education','Educational','Libraries&Demo','Word']

entertainment = ['Entertainment','Action','Adventure','Arcade','Board','Comics','Card','Casino','Casual','Events',
                 'Puzzle','Racing','Role Playing','Simulation','Sports','Strategy','Trivia']
```

```python
df = df.loc[(df['Category'].isin(education)) | (df['Category'].isin(entertainment))]
```

```python
appId = pd.DataFrame(columns=['App Id', 'description'])
appId['App Id']= df['App Id'].unique()
```

```python
from google_play_scraper import app

for index, row in appId[:10].iterrows():
    try:
        value = row['App Id']
        result = app(value)
        appId.loc[index,'description']=result['description']
    except:
        appId.loc[index,'description']='Not Found'
```
